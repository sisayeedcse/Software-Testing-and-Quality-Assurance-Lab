# TASK 1

    Calculates the bonus points and discount for a customer.
    
    Rules:
    1. If purchase > $500, they get a 10% discount.
    2. If they've been a member for 5+ years, they get an extra 5% discount.
    3. Priority members get a flat 50 bonus points.
    4. If purchase is under $100 and NOT priority, they get 0 bonus points.




1.  Draw CFG
2.  Calculate Cyclomatic Complexity
3.  Write down the Independent paths
4. Write down the test cases
5. Find the bugs



```python
def calculate_loyalty_bonus(purchase_amount, membership_years, is_priority_member):

    discount = 0.0
    bonus_points = 10

    # Branch 1: Purchase Discount
    if purchase_amount > 500:
        discount = 0.10

    # Branch 2: Seniority Discount
    if membership_years >= 5:
        discount = discount + 0.05

    # Branch 3: Priority Status
    if is_priority_member:
        bonus_points = bonus_points + 50
    else:
        # Branch 4: Low-spend non-priority logic
        if purchase_amount < 100:
            bonus_points = 5 # BUG: Should be 0 per rules

    # Branch 5: Final calculation check
    if discount > 0.15:
        # This should technically be impossible based on the rules,
        # but let's see if their path testing finds a way here.
        bonus_points = bonus_points * 2

    return discount, bonus_points
```

```python
calculate_loyalty_bonus(200, 1, 0)
```

```python
def evaluate_loan(age, income, score, co_signers):
    rate = 0.07
    approved = False
    processed_co_signers = 0

    if age < 18:
        print("Denied: Underage")
        return

    if income > 100000:
        rate -= 0.02

    if score < 600:
        while processed_co_signers < co_signers:
            if score > 700:
                approved = True
            processed_co_signers += 1
    else:
        approved = True

    if rate < 0.02:
        rate = 0.01

    if approved:
        print(f"Approved at {rate}")
```

```python
evaluate_loan(10, 10000, 544, 1)
```

```python
def evaluate_insurance_claim(age, claim_amount, policy_years, previous_claims, has_documentation):
    coverage = 1.0
    claim_approved = 0
    verification_level = 0
    processed_claims = 0
    print("Processing claim...")

    if age < 21 or age > 80:
        print("Denied: Age out of range")
        return

    if policy_years >= 5:
        coverage += 0.15
        verification_level = 1

    if claim_amount > 50000:
        verification_level = 2
        if policy_years < 2:
            claim_approved = 1
    else:
        verification_level = 1

    while processed_claims < previous_claims:
        coverage -= 0.05

        if coverage < 0.5:
            coverage = 0.5
        processed_claims += 1

    if has_documentation == 1:
        claim_approved = 1

    if coverage > 2.0:
        coverage = 2.0
    if claim_approved:
        print(f"Claim approved with {coverage * 100:.2f}% coverage")
    else:
        print("Claim requires manual review")


# Test cases for students to verify their path coverage
if __name__ == "__main__":
    print("Test Case 1: Age out of range")
    evaluate_insurance_claim(17, 30000, 3, 0, 1)
    print("\n" + "="*50 + "\n")
```
